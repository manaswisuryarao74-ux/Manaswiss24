```python


years = ['2001', '2006', '2010', '2015', '2018']
sachin_runs = [1600, 1300, 1562, 0, 0]
kohli_runs = [0, 0, 995, 1380, 1325]
sehwag_runs = [850, 900, 1100, 0, 0]

x = np.arange(len(years))
width = 0.25

mp.figure(figsize = (10,6))
mp.bar(x - width, sachin_runs, width, label = 'Sachin')
mp.bar(x,kohli_runs, width, label = 'Kohli')
mp.bar(x + width, sehwag_runs, width, label = 'Sehwag')

mp.xlabel('YEARS')
mp.ylabel('RUNS')
mp.title('Over 5 selected Years')
mp.xticks(x, years)

mp.legend()
mp.show()
```

Q.NO-5.  
Bar Chart with Annotations: Draw a vertical bar chart for five categories of your choice. Annotate the value on top of each bar.

```python
import matplotlib.pyplot as mp

Subjects = ['Math', 'Marathi', 'Hindi', 'English', 'Biology']
Markes = [45, 25, 30, 20, 15]
colours = ['#FBFE52','#52FE85','#FE52EA','#FE525D','#FE8852']

mp.figure(figsize=(8, 6))
bars = mp.bar(Subjects, Markes, color=colours)

for bar in bars:
    height = bar.get_height()
    mp.text(bar.get_x() + bar.get_width()/2, height + 1, str(height),
             ha='center', va='bottom', fontsize=10)


mp.title("Subject's Markes")
mp.xlabel("SUBJECTS")
mp.ylabel('MARKES')

mp.grid(axis='y', linestyle='--', alpha=0.7)

mp.show()
```

Q.NO-6.  
Multiple Lines in One Plot: Plot three mathematical functions (sin, cos, exp) on the same axes, each with different colors and line styles. Add a legend.

```python
import matplotlib.pyplot as mp
import numpy as np

x = np.linspace(0, 2 * np.pi, 100)

y_sin = np.sin(x)
y_cos = np.cos(x)
y_exp = np.exp(x/4)

mp.figure(figsize = (8,6))
mp.plot(x, y_sin, color = '#80F980', linestyle = '-', label = 'sin(x)')
mp.plot(x, y_cos, color = '#80F4F9', linestyle = '--', label = 'cos(x)')
mp.plot(x, y_exp, color = 'red', linestyle = ':', label = 'exp(x/4)')

mp.xlabel("X")
mp.ylabel("Y")
mp.title("mathematical functions (sin, cos, exp)")
mp.grid(True)
mp.legend()

mp.show()
```

Q.NO-8.  
Subplots Grid: Create a 2x2 grid of subplots: line plot, bar chart, scatter plot, and histogram, each in a different subplot. Add an overall(supertitle) for the figure.

```python
import matplotlib.pyplot as mp
import numpy as np

x_line = np.linspace(0, 10, 100)
y_line = np.sin(x_line)

Cars = ['Yamaha', 'BMW', 'Royal Enfieldd', 'KTM', 'Ducati']
networth = [95, 85, 7.5, 82, 28]

x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)

hist_data = [96, 82, 75, 91, 67, 88, 59, 3, 84, 62, 
        90, 77, 68, 94, 56, 7, 7, 91, 70, 85, 
        92, 66, 74, 0, 97, 94, 60, 76, 83, 
        93, 78, 71, 58, 86, 72, 3, 69, 65, 98, 
        57, 99, 55, 81, 53, 52, 54, 1, 50, 100]

fig, axs = mp.subplots(2, 2, figsize=(12, 10))

# Line plot
axs[0, 0].plot(x_line, y_line, color='#CECECE')
axs[0, 0].set_title("Line Plot")

# Bar chart
axs[0, 1].bar(Cars, networth, color='red')
axs[0, 1].set_title("Cars Networth in Billion $ :(Bar Chart)")

# Scatter plot
axs[1, 0].scatter(x_scatter, y_scatter, color='#A4FF4F',s=50)
axs[1, 0].set_title("Scatter Plot")

# Histogram
axs[1, 1].hist(hist_data, bins=5, color='#FE9900', edgecolor='black')
axs[1, 1].set_title("Histogram")

# Supertitle
fig.suptitle("2x2 Grid of Various Plots", fontsize=16)

# Adjust layout to prevent overlap
fig.tight_layout(rect=[0, 0.03, 1, 0.95])

# Show the plot
mp.show()
```

Q.NO-9.    
Stack Plot: Plot the cumulative sales of 3 products over 12 months as a stack plot using made-up data.

```python
import matplotlib.pyplot as mp
import numpy as np

months = np.arange(1, 13)

product_A = [12, 17, 22, 20, 24, 27, 31, 33, 37, 41, 44, 48]
product_B = [6, 9, 13, 11, 13, 18, 21, 23, 26, 28, 30, 32]
product_C = [3, 5, 7, 8, 9, 11, 13, 14, 15, 17, 19, 21]

mp.figure(figsize=(9, 5))
mp.stackplot(months, product_A, product_B, product_C,
              labels=['Android', 'iOS', 'Linux'],
              colors=['#C3FC8D', '#FCB28D', '#8DFCDA'])

mp.xlabel("Month")
mp.ylabel("Sales")
mp.title("Cumulative Sales of 3 Products Over 12 Months")
mp.legend(loc='upper left')
mp.grid(True, linestyle='--', alpha=0.7)

mp.show()
```

Q.NO-12.   
Axis and Legend Customization: For a random line plot, reverse the x-axis, move the 
legend outside the plot, and increase legend font size.

```python
import matplotlib.pyplot as mp
import numpy as np

x = np.arange(0, 10, 1) #0-9
y1 = np.random.randint(10, 50, size=10)
y2 = np.random.randint(20, 60, size=10)

# Plot lines
mp.plot(x, y1, label='Ducati', color='#0F0F0F')
mp.plot(x, y2, label='Astin Martin', color='#F90404')

# Reverse x-axis
mp.gca().invert_xaxis() # gca=gets current axis

mp.title("Random Line Plot with Custom Axis and Legend")
mp.xlabel("X-axis")
mp.ylabel("Y-axis")

# Move legend outside the plot (right side) & increase font size
mp.legend(loc='center left', bbox_to_anchor=(1, 0.9), fontsize='large')
mp.tight_layout()
mp.show()
```
